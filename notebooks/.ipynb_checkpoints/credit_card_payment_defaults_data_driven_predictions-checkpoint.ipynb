# Credit Card Payment Defaults: Data-Driven Predictions

This project focuses on building a classification model to predict whether a credit card client will default on their payment next month using the UCI credit card dataset. Key steps include:

**Data Exploration:** Analyze and visualize the data to understand variable distributions.

**Preprocessing:** Build a Scikit-learn pipeline for transformations, data splitting (80% training, 20% testing), and 5-fold cross-validation.

**Classification:** Compare Random Forest and KNN classifiers using GridSearchCV to optimize hyperparameters and evaluate performance via ROC-AUC.

The project identifies key features like repayment status, credit amount, age, and sex to build an accurate and efficient model for predicting defaults.

```python
def check_data_balance(series, style='seaborn-pastel'):
    with plt.style.context(style):
        unique = series.value_counts()
        # to show unique value counts of the target:
        display(unique)
        # to plot a pie chart for the target to see if the data are balanced:
        plt.pie(unique, explode=[0.05]*len(unique), labels=unique.index, autopct='%1.1f%%');
```

- `PAY_AMT3`: Amount of previous payment in July, 2005 (NT dollar)

There should be an equal number of samples from positive and negative categories in a balanced dataset. As in our example, if the samples from one of the classes outnumber those from the other, this results in the data being skewed in favour of one of the classes. Training may be problematic, as the majority class can dominate the minority, and the classifier may only learn one concept instead of two. Even so, our case doesn't appear to be very severe and we may be able to handle it without much effort.

Let's define our pipeline:

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, FunctionTransformer, OneHotEncoder
```

```python
def assign_others_education(val, valid_values):
    mask = np.isin(val, valid_values)
    val[mask==False] = 4
    return pd.DataFrame(val)
```

```python
categorical_pipeline = Pipeline([
    ('ohe', OneHotEncoder(handle_unknown='ignore')),
])
numeric_pipeline = Pipeline([
    ('mms', MinMaxScaler(feature_range=(0,1))),
])
full_pipeline = ColumnTransformer([
    ('unk_edu', FunctionTransformer(func=assign_others_education,
                                    kw_args={'valid_values': [1, 2, 3, 4]}), 'EDUCATION'),
    ('categorical', categorical_pipeline, categorical_X),
    ('numeric', numeric_pipeline, numeric_X),
])
```

Let's apply our pipeline:

```python
full_pipeline.fit(X_train, y_train)
X_train_transformed = full_pipeline.transform(X_train)
X_test_transformed = full_pipeline.transform(X_test)
print(f'X_train_transformed.shape: {X_train_transformed.shape}')
print(f'X_test_transformed.shape: {X_test_transformed.shape}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
num_folds = 5
seed = 123
k_fold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)
```

```python
def display_scores(scores):
    print('Scores:', scores)
    print('Mean:', scores.mean())
    print('Standard deviation:', scores.std())
```

```python
rfc_default = RandomForestClassifier(random_state=seed)
rfc_default.fit(X_train_transformed, y_train)
rfc_default_scores = cross_val_score(rfc_default, X_train_transformed, y_train, scoring='roc_auc', cv=k_fold)
display_scores(rfc_default_scores)
```

```python
knc_default = KNeighborsClassifier()
knc_default.fit(X_train_transformed, y_train)
knc_default_scores = cross_val_score(knc_default, X_train_transformed, y_train, scoring='roc_auc', cv=k_fold)
display_scores(knc_default_scores)
```

```python
knc_20_neighbors = KNeighborsClassifier(n_neighbors=20)
knc_20_neighbors.fit(X_train_transformed, y_train)
knc_20_neighbors_scores = cross_val_score(knc_20_neighbors, X_train_transformed, y_train,
                                          scoring='roc_auc', cv=k_fold)
display_scores(knc_20_neighbors_scores)
```

```python
from sklearn.metrics import roc_curve
from sklearn.model_selection import cross_val_predict
```

```python
y_probas_rfc_default = cross_val_predict(rfc_default, X_train_transformed, y_train,
                                         cv=5, method='predict_proba')
y_scores_rfc_default = y_probas_rfc_default[:, 1]
fpr_rfc_default, tpr_rfc_default, thresholds_rfc_default = roc_curve(y_train, y_scores_rfc_default)
```

```python
y_probas_knc_20_neighbors = cross_val_predict(knc_20_neighbors, X_train_transformed, y_train,
                                              cv=5, method='predict_proba')
y_scores_knc_20_neighbors = y_probas_knc_20_neighbors[:, 1]
fpr_knc_20_neighbors, tpr_knc_20_neighbors, thresholds_knc_20_neighbors = roc_curve(y_train,
                                                                                    y_scores_knc_20_neighbors)
```

```python
gmeans_rfc_default = np.sqrt(tpr_rfc_default * (1 - fpr_rfc_default))
index_rfc_default = np.argmax(gmeans_rfc_default)
print('Random Forest Classifier, optimal threshold =',
      thresholds_rfc_default[index_rfc_default])
```

```python
gmeans_knc_20_neighbors = np.sqrt(tpr_knc_20_neighbors * (1 - fpr_knc_20_neighbors))
index_knc_20_neighbors = np.argmax(gmeans_knc_20_neighbors)
print('K-Nearest Neighbors Classifier, optimal threshold =',
      thresholds_knc_20_neighbors[index_knc_20_neighbors])
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr_rfc_default, tpr_rfc_default, linewidth=2, label='Random Forest classifier')
plt.plot(fpr_knc_20_neighbors, tpr_knc_20_neighbors, linewidth=2, label='K-Nearest Neighbors classifier')
plt.scatter(fpr_rfc_default[index_rfc_default], tpr_rfc_default[index_rfc_default],
            marker='o', color='black')
plt.scatter(fpr_knc_20_neighbors[index_knc_20_neighbors], tpr_knc_20_neighbors[index_knc_20_neighbors],
            marker='o', color='black', label='Optimal threshold')
plt.plot([0, 1], [0, 1], 'k--') 
plt.axis([0, 1, 0, 1])                                   
plt.xlabel('False Positive Rate (1 - Specificity)') 
plt.ylabel('True Positive Rate (Sensitivity)')   
plt.legend()
plt.grid(True)
```

Comparing ROC curves: the Random Forest classifier is superior to the K-Nearest Neighbors classifier because its ROC curve is closer to the top-left corner, and it has a greater AUC (according to the results of cross-validation, the ROC AUC sore for the Random Forest classifier with default parameters (i.e., n_estimators=100) is the highest, 0.763+/-0.006).
